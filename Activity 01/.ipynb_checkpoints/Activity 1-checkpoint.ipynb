```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin, cos, pi, sqrt

# SINUSOID

N = 400

x = np.linspace(-1,1,N)
y = sin(8*pi*x)    # sin(2k*pi x) k determines frequency per unit

X,Y = np.meshgrid(x,y)
sinusoid = np.transpose(Y)

plt.imshow(sinusoid, cmap='gray')
# plt.imsave('sinusoid.jpg',sinusoid,cmap='gray')
# plt.imsave('sinusoid.png',sinusoid,cmap='gray')
# plt.imsave('sinusoid.bmp',sinusoid,cmap='gray')
# plt.imsave('sinusoid.tiff',sinusoid,cmap='gray')
```

```python
# GRATING

x = np.ones(N)
x *= -1

ctr = 0
div = 10
    
for j in range(int(N/div)):
    for i in range(ctr*div,N):
        x[i] *= -1
    ctr += 1

y = x
        
X,Y = np.meshgrid(x,y)

plt.imshow(X, cmap='gray')
# plt.imsave('gratings.jpg',X,cmap='gray')
# plt.imsave('gratings.png',X,cmap='gray')
# plt.imsave('gratings.bmp',X,cmap='gray')
# plt.imsave('gratings.tiff',X,cmap='gray')
```

```python
# ANNULUS

N = 2000 # the higher num is the finer
x = np.linspace(-1,1,num = N)
y = x
X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2)
A = np.zeros(np.shape(R))
A[np.where(R<0.9)] = 1.0       # set radius of 0.9 to 1 to create a disk
A[np.where(R<0.2)] = 0.0       # set radius of 0.2 back to 0 to punch a hole in the middle

plt.imshow(A, cmap='gray_r')
# plt.imsave('annulus.jpg',A,cmap='gray')
# plt.imsave('annulus.png',A,cmap='gray')
# plt.imsave('annulus.bmp',A,cmap='gray')
# plt.imsave('annulus.tiff',A,cmap='gray')
```

```python
# JWST

def space(plots, occupied):
    coord = plots        
    colors = ["white" if k == 0 else "yellow" for k in occupied.values()]
    
    # Horizontal coordinates array
    hcoord = [c[0] for c in coord]                              

    # Vertical coordinates array
    vcoord = [2.*sin(radians(60))*(c[1]-c[2])/3. for c in coord]
                                            

    fig, ax = plt.subplots(1, figsize=(8,8))
    ax.set_aspect('equal')                  

    
    for i, j, k in zip(hcoord, vcoord, colors):
        color = k[0]                                                       
        hex = RegularPolygon((i, j), numVertices=6, radius=2./3,           
                             orientation=radians(30), facecolor = color,   
                             alpha=0.3, edgecolor='k')
        ax.add_patch(hex)    


    ax.scatter(hcoord, vcoord, alpha=0)
    
    plt.axis('off')
#     plt.savefig('jwst.png')
#     plt.savefig('jwst.jpg')
# #     plt.savefig('jwst.bmp')
#     plt.savefig('jwst.tiff')
    plt.show()
```

```python
from numpy import sin, cos, pi, sqrt

# OLYMPIC LOGO

N = 500
x = np.linspace(-10,10,N)
y = x
X,Y = np.meshgrid(x,y)
Rd = np.ones((N,N))
Gn = np.ones((N,N))
Bl = np.ones((N,N))

# draw colored circles
Rt,Rc,Ri,deg = 2,3,2.5,30

xt,yt = Rt*cos(deg*pi/180), Rt*sin(deg*pi/180)

# TOP MIDDLE (BLACK)
R = sqrt((X)**2 + (Y+Rt)**2)

Rd[np.where(R<Rc)] = 0.0    # RGB of BLACK ring
Gn[np.where(R<Rc)] = 0.0
Bl[np.where(R<Rc)] = 0.0

Rd[np.where(R<Ri)] = 1.0    # white inside
Gn[np.where(R<Ri)] = 1.0
Bl[np.where(R<Ri)] = 1.0

# TOP RIGHT (RED)
R = sqrt((X-4*xt)**2 + (Y+Rt)**2)

Rd[np.where(R<Rc)] = 1.0    # RGB of RED ring
Gn[np.where(R<Rc)] = 0.0
Bl[np.where(R<Rc)] = 0.0

Rd[np.where(R<Ri)] = 1.0    # white inside
Gn[np.where(R<Ri)] = 1.0
Bl[np.where(R<Ri)] = 1.0

# TOP LEFT (BLUE)
R = sqrt((X+4*xt)**2 + (Y+Rt)**2)

Rd[np.where(R<Rc)] = 0.0    # RGB of BLUE ring
Gn[np.where(R<Rc)] = 0.0
Bl[np.where(R<Rc)] = 1.0

Rd[np.where(R<Ri)] = 1.0    # white inside
Gn[np.where(R<Ri)] = 1.0
Bl[np.where(R<Ri)] = 1.0

# BOTTOM RIGHT (GREEN)
R = sqrt((X-2*xt)**2 + (Y-yt)**2)

Rd[np.where(R<Rc)] = 0.0    # RGB of GREEN ring
Gn[np.where(R<Rc)] = 1.0
Bl[np.where(R<Rc)] = 0.0

Rd[np.where(R<Ri)] = 1.0    # white inside
Gn[np.where(R<Ri)] = 1.0
Bl[np.where(R<Ri)] = 1.0

# BOTTOM LEFT (YELLOW)
R = sqrt((X+2*xt)**2 + (Y-yt)**2)

Rd[np.where(R<Rc)] = 1.0    # RGB of YELLOW ring
Gn[np.where(R<Rc)] = 1.0
Bl[np.where(R<Rc)] = 0.0

Rd[np.where(R<Ri)] = 1.0    # white inside
Gn[np.where(R<Ri)] = 1.0
Bl[np.where(R<Ri)] = 1.0

# OVERLAP TOP LEFT - BOTTOM LEFT (BLUE-YELLOW)
R1 = sqrt((X+4*xt)**2 + (Y+Rt)**2)
R2 = sqrt((X+2*xt)**2 + (Y-yt)**2)

for i in range(N):              # check if every pixel is between the gap of BLUE and YELLOW ring
    for j in range(N):
        if R1[i,j] < Rc and R2[i,j] < Ri:
            Rd[i,j] = 0.0
            Gn[i,j] = 0.0
            Bl[i,j] = 1.0

for i in range(N):              # set the actual gap between BLUE and YELLOW to white background
    for j in range(N):
        if R1[i,j] < Ri and R2[i,j] < Ri:
            Rd[i,j] = 1.0
            Gn[i,j] = 1.0
            Bl[i,j] = 1.0
            
# OVERLAP TOP MIDDLE - BOTTOM LEFT (BLACK-YELLOW)
R1 = sqrt((X)**2 + (Y+Rt)**2)
R2 = sqrt((X+2*xt)**2 + (Y-yt)**2)

for i in range(N):              # check if every pixel is between the gap of BLACK and YELLOW ring
    for j in range(N):
        if R1[i,j] < Rc and R2[i,j] < Ri:
            Rd[i,j] = 0.0
            Gn[i,j] = 0.0
            Bl[i,j] = 0.0

for i in range(N):              # set the actual gap between BLACK and YELLOW to white background
    for j in range(N):
        if R1[i,j] < Ri and R2[i,j] < Ri:
            Rd[i,j] = 1.0
            Gn[i,j] = 1.0
            Bl[i,j] = 1.0
            
# OVERLAP TOP MIDDLE - BOTTOM RIGHT (BLACK-GREEN)
R1 = sqrt((X)**2 + (Y+Rt)**2)
R2 = sqrt((X-2*xt)**2 + (Y-yt)**2)

for i in range(N):              # check if every pixel is between the gap of BLACK and GREEN ring
    for j in range(N):
        if R1[i,j] < Rc and R2[i,j] < Ri:
            Rd[i,j] = 0.0
            Gn[i,j] = 0.0
            Bl[i,j] = 0.0

for i in range(N):              # set the actual gap between BLACK and GREEN to white background
    for j in range(N):
        if R1[i,j] < Ri and R2[i,j] < Ri:
            Rd[i,j] = 1.0
            Gn[i,j] = 1.0
            Bl[i,j] = 1.0
            
# OVERLAP TOP RIGHT - BOTTOM RIGHT (RED-GREEN)
R1 = sqrt((X-4*xt)**2 + (Y+Rt)**2)
R2 = sqrt((X-2*xt)**2 + (Y-yt)**2)

for i in range(N):
    for j in range(N):
        if R1[i,j] < Rc and R2[i,j] < Ri:    # check if every pixel is between the gap of RED and GREEN ring
            Rd[i,j] = 1.0
            Gn[i,j] = 0.0
            Bl[i,j] = 0.0

for i in range(N):              # set the actual gap between RED and GREEN to white background
    for j in range(N):
        if R1[i,j] < Ri and R2[i,j] < Ri:
            Rd[i,j] = 1.0
            Gn[i,j] = 1.0
            Bl[i,j] = 1.0

I = np.zeros((N,N,3))

I[...,0] = Rd
I[...,1] = Gn
I[...,2] = Bl

fig = plt.figure()
plt.imshow(I)
# plt.imsave('olympics.jpg',I)
# plt.imsave('olympics.png',I)
# plt.imsave('olympics.bmp',I)
# plt.imsave('olympics.tiff',I)
```

```python
# HISTOGRAM BACKPROJECTION
def mean_gray(img):
    grayscale = np.dot(img[...,:3],[0.2989, 0.5870, 0.1140])
    return grayscale

dark_img = plt.imread('dark_image.jpg')
dark_img_gray = mean_gray(dark_img)

plt.imshow(dark_img_gray,cmap='gray')
```

```python
dark_img_gray_norm = 255*dark_img_gray/dark_img_gray.max()

plt.figure(figsize=(15,5))

plt.subplot(121)
dark_img_gray_PDF, bins = np.histogram(dark_img_gray_norm, bins=256, density=True)    # normalized histogram
plt.xlim(0,255)
plt.title('PDF of original image')
plt.plot(bins[:-1], dark_img_gray_PDF)

cumsum = np.cumsum(dark_img_gray_PDF)
plt.subplot(122)
plt.xlim(0,255)
plt.title('CDF of riginal image')
dark_img_gray_CDF = plt.plot(bins[:-1],cumsum)

plt.show()
```

```python
# desired CDF (Linear)

desired_x = np.arange(0,256)
desired_y = np.linspace(0,1,256)

plt.title("Desired CDF (Linear)")
plt.xlim(0,255)
plt.ylim(0,1)
plt.plot(desired_x, desired_y)
plt.savefig('linear.png')
plt.show()
```

```python
dark_img_gray_int = dark_img_gray.astype(int)

old_x = desired_x
new_x = cumsum*255

dark_img_gray_flat = dark_img_gray_int.ravel()

index = np.searchsorted(old_x, dark_img_gray_flat)

dark_img_gray_flat[:] = new_x[index]
```

```python
plt.figure(figsize=(15,5))

plt.subplot(121)
plt.title('Original image')
plt.imshow(dark_img_gray, cmap='gray')

plt.subplot(122)
plt.title('Histogram equalized image (Linear)')
plt.imshow(dark_img_gray_int, cmap = 'gray')
plt.savefig('dark_image_enhanced_linear.png')
plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(121)
dark_img_eqlzd_PDF, bins = np.histogram(dark_img_gray_int, bins=256, density=True)    # normalized histogram
plt.xlim(0,255)
plt.title('PDF of histogram equalized image')
plt.plot(bins[:-1], dark_img_eqlzd_PDF)

cumsum = np.cumsum(dark_img_eqlzd_PDF)
plt.subplot(122)
plt.xlim(0,255)
plt.title('CDF of histogram equalized image')
dark_img_gray_CDF = plt.plot(bins[:-1],cumsum)
plt.savefig('dark_image_enhanced_linear_PDF_CDF.png')

plt.show()
```

```python
# desired CDF (Sigmoid)

def sigmoid(x):
    return 1 / (1 + np.exp((med-x)/15))

def inverse_sigmoid(y):
    return med - 15*np.log(abs((1/y)-1))

sigmoid_x = np.arange(0,256)
med = np.median(sigmoid_x)
sigmoid_y = sigmoid(sigmoid_x)

plt.title("Desired CDF (Sigmoid)")
plt.xlim(0,255)
plt.ylim(0,1)
plt.plot(sigmoid_x, sigmoid_y)
plt.savefig('sigmoid.png')
plt.show()
```

```python
dark_img_gray_norm = 255*dark_img_gray/dark_img_gray.max()

dark_img_gray_PDF, bins = np.histogram(dark_img_gray_norm, bins=256, density=True)

cumsum = np.cumsum(dark_img_gray_PDF)
```

```python
dark_img_gray_int = dark_img_gray.astype(int)

old_x = desired_x
new_x = inverse_sigmoid(cumsum)

mask = np.in1d(dark_img_gray_int,old_x)

index = np.searchsorted(old_x,dark_img_gray_int.ravel()[mask])

dark_img_gray_int.ravel()[mask] = new_x[index]
```

```python
# Same plotting parameters
plt.figure(figsize = (15,5))


plt.subplot(121)
plt.title("Original image")
plt.imshow(dark_img_gray, cmap = 'gray')

plt.subplot(122)
plt.title('Histogram equalized image (Sigmoid)')
plt.imshow(dark_img_gray_int, cmap = 'gray')
plt.savefig('dark_image_enhanced_sigmoid.png')

plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(121)
dark_img_eqlzd_PDF, bins = np.histogram(dark_img_gray_int, bins=256, density=True)    # normalized histogram
plt.title('PDF of histogram equalized image')
plt.plot(bins[:-1], dark_img_eqlzd_PDF)

cumsum = np.cumsum(dark_img_eqlzd_PDF)
plt.subplot(122)
plt.title('CDF of histogram equalized image')
dark_img_eqlzd_CDF = plt.plot(cumsum)
plt.savefig('dark_image_enhanced_sigmoid_PDF_CDF.png')

plt.show()
```

```python
# CONTRAST STRECTCHING
def mean_gray(img):
    grayscale = np.dot(img[...,:3],[0.2989, 0.5870, 0.1140])
    return grayscale

dark_img = plt.imread('dark_image.jpg')
dark_img_gray = mean_gray(dark_img)

plt.imshow(dark_img_gray,cmap='gray')
```

```python
dark_img_gray_norm = 255*dark_img_gray/dark_img_gray.max()

binss = np.arange(0,256)

plt.figure(figsize=(15,5))

plt.subplot(121)
dark_img_gray_PDF, bins = np.histogram(dark_img_gray_norm, bins=256, density=True)    # normalized histogram
plt.xlim(0,255)
plt.title('PDF of original image')
plt.plot(binss, dark_img_gray_PDF)

dark_img_gray_CDF = np.cumsum(dark_img_gray_PDF)
plt.subplot(122)
plt.xlim(0,255)
plt.title('CDF of original image')
plt.plot(binss,dark_img_gray_CDF)
plt.savefig('contrast_stretch_orig_PDF_CDF.png')
plt.show()
```

```python
def contrast_stretch(img_gray,min_percentile,max_percentile):
    img_gray_norm = 255*img_gray/img_gray.max()
    img_gray_PDF, bins = np.histogram(img_gray, bins=np.arange(0,256), density=True)
    img_gray_CDF = np.cumsum(img_gray_PDF)
    
    mins = np.where(img_gray_CDF <= min_percentile)
    I_min = mins[0].max()
    
    maxs = np.where(img_gray_CDF >= max_percentile)
    I_max = maxs[0].min()
    
    I = (img_gray - I_min)/(I_max - I_min)*255
    
    return I
```

```python
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.imshow(dark_img_gray,cmap='gray')
plt.title('Original image')

dark_img_gray_stretched = contrast_stretch(dark_img_gray,0.3,0.6)
plt.subplot(122)
plt.imshow(dark_img_gray_stretched,cmap='gray')
plt.title('Contrast stretched image')
# plt.savefig('contrast_stretch.png')
plt.show()
```

```python
bins = np.arange(0,256)
plt.figure(figsize=(15,5))

plt.subplot(121)
dark_img_gray_stretched_PDF, bins = np.histogram(dark_img_gray_stretched, bins=bins, density=True)    # normalized histogram
plt.title('PDF of contrast stretched image')
plt.plot(dark_img_gray_stretched_PDF)

dark_img_gray_stretched_CDF = np.cumsum(dark_img_gray_stretched_PDF)
plt.subplot(122)
plt.title('CDF of stretched image')
plt.plot(dark_img_gray_stretched_CDF)
plt.savefig('contrast_stretch_new_PDF_CDF.png')
plt.show()
```

```python
percentiles = [[0.0001,1],[0.1,0.9],[0.15,0.8],[0.2,0.7],[0.25,0.6]]
colors = ['r','orange','y','g','b']
CDF_array = []

bins = np.arange(0,256)

for i in range(len(percentiles)):
    print('min: {}, max: {}'.format(percentiles[i][0],percentiles[i][1]))
    
    dark_img_gray_stretched = contrast_stretch(dark_img_gray,percentiles[i][0],percentiles[i][1])
    dark_img_gray_stretched = np.reshape(dark_img_gray_stretched, (-1,1))    
    plt.figure(figsize=(15,5))

    plt.subplot(121)
    dark_img_gray_stretched_PDF = plt.hist(dark_img_gray_stretched, bins=bins, density=True)
    plt.title('PDF of contrast stretched image')
    
    dark_img_gray_stretched_CDF = np.cumsum(dark_img_gray_stretched_PDF[0])
    dark_img_gray_stretched_CDF = np.reshape(dark_img_gray_stretched_CDF,(-1,1))
    CDF_array.append(dark_img_gray_stretched_CDF)
    plt.subplot(122)
    plt.title('CDF of stretched image')
    plt.plot(dark_img_gray_stretched_CDF, color=colors[i])
    
    plt.show()

for i in range(len(percentiles)):
    plt.plot(CDF_array[i],label=percentiles[i], color=colors[i])
    
plt.legend()
plt.title('CDF of stretched image')
plt.savefig('contrast_stretch_plot.png')
plt.show()
```

## Gray World Algorithm

```python
def gray_world(img):

    img = plt.imread('room.jpg')
    img_r = img[:,:,0]
    img_g = img[:,:,1]
    img_b = img[:,:,2]

    r_ave = img_r.mean()
    g_ave = img_g.mean()
    b_ave = img_b.mean()

    r_gw = img_r/r_ave
    g_gw = img_g/g_ave
    b_gw = img_b/b_ave
    
    r_gw = np.clip(r_gw, 0, 1)
    g_gw = np.clip(g_gw, 0, 1)
    b_gw = np.clip(b_gw, 0, 1)

    img_gw = np.dstack([r_gw,g_gw,b_gw])
    
    return img_gw

room_gw = gray_world(room)

# plt.figure(figsize=(15,5))
# plt.subplot(121)
# plt.imshow(room)
# plt.subplot(122)
plt.imshow(room_gw)
plt.imsave('room_grayworld.png',room_gw)
# plt.show()
```

```python
room_white_patch = room[0:90, 100:200, :]

room_white_patch_r = room_white_patch[:, :, 0]
room_white_patch_g = room_white_patch[:, :, 1]
room_white_patch_b = room_white_patch[:, :, 2]

white_r_ave = room_white_patch_r.mean()
white_g_ave = room_white_patch_g.mean()
white_b_ave = room_white_patch_b.mean()

room_wb_r = room[:, :, 0] / white_r_ave
room_wb_g = room[:, :, 1] / white_g_ave
room_wb_b = room[:, :, 2] / white_b_ave

room_wb_r = np.clip(room_wb_r, 0, 1)
room_wb_g = np.clip(room_wb_g, 0, 1)
room_wb_b = np.clip(room_wb_b, 0, 1)

room_wb = np.dstack([room_wb_r, room_wb_g, room_wb_b])

plt.figure(figsize=(15,5))
plt.subplot(121)
plt.imshow(room)
plt.subplot(122)
plt.imshow(room_wb)
plt.show()

plt.imsave('room_whitepatch.png',room_wb)
```
