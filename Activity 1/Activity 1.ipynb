```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin, cos, pi

N = 400

x = np.linspace(-1,1,N)
y = sin(8*pi*x)

plt.plot(x,y)
```

```python
x = np.zeros(N)
for i in range(N):
    if i%2 == 0:
        x[i]+=1
        
y = x
        
X,Y = np.meshgrid(x,y)

plt.imshow(X, cmap='gray')
```

```python
N = 2000 #the higher num is the finer
x = np.linspace(-1,1,num = N)
y = x
X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2)
A = np.zeros(np.shape(R))
A[np.where(R<0.9)] = 1.0       # set radius of 0.9 to 1
A[np.where(R<0.2)] = 0.0       # set radius of 0.2 back to 0 to punch a hole in the middle

plt.imshow(A, cmap='gray_r')
```
