```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin, cos, pi, sqrt

# SINUSOID

N = 400

x = np.linspace(-1,1,N)
y = sin(8*pi*x)    # sin(2k*pi x) k determines frequency per unit

X,Y = np.meshgrid(x,y)
sinusoid = np.transpose(Y)

plt.imshow(sinusoid, cmap='gray')
plt.imsave('sinusoid.jpg',sinusoid,cmap='gray')
plt.imsave('sinusoid.png',sinusoid,cmap='gray')
plt.imsave('sinusoid.bmp',sinusoid,cmap='gray')
plt.imsave('sinusoid.tiff',sinusoid,cmap='gray')
```

```python
# GRATING

x = np.ones(N)
x *= -1

ctr = 0
div = 10
    
for j in range(int(N/div)):
    for i in range(ctr*div,N):
        x[i] *= -1
    ctr += 1

y = x
        
X,Y = np.meshgrid(x,y)

plt.imshow(X, cmap='gray')
plt.imsave('gratings.jpg',X,cmap='gray')
plt.imsave('gratings.png',X,cmap='gray')
plt.imsave('gratings.bmp',X,cmap='gray')
plt.imsave('gratings.tiff',X,cmap='gray')
```

```python
# ANNULUS

N = 2000 # the higher num is the finer
x = np.linspace(-1,1,num = N)
y = x
X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2)
A = np.zeros(np.shape(R))
A[np.where(R<0.9)] = 1.0       # set radius of 0.9 to 1 to create a disk
A[np.where(R<0.2)] = 0.0       # set radius of 0.2 back to 0 to punch a hole in the middle

plt.imshow(A, cmap='gray_r')
plt.imsave('annulus.jpg',A,cmap='gray')
plt.imsave('annulus.png',A,cmap='gray')
plt.imsave('annulus.bmp',A,cmap='gray')
plt.imsave('annulus.tiff',A,cmap='gray')
```

```python
# JWST

def space(plots, occupied):
    coord = plots        
    colors = ["white" if k == 0 else "yellow" for k in occupied.values()]
    
    # Horizontal coordinates array
    hcoord = [c[0] for c in coord]                              

    # Vertical coordinates array
    vcoord = [2.*sin(radians(60))*(c[1]-c[2])/3. for c in coord]
                                            

    fig, ax = plt.subplots(1, figsize=(8,8))
    ax.set_aspect('equal')                  

    
    for i, j, k in zip(hcoord, vcoord, colors):
        color = k[0]                                                       
        hex = RegularPolygon((i, j), numVertices=6, radius=2./3,           
                             orientation=radians(30), facecolor = color,   
                             alpha=0.3, edgecolor='k')
        ax.add_patch(hex)    


    ax.scatter(hcoord, vcoord, alpha=0)
    
    plt.axis('off')
    plt.show()
#     plt.savefig('jwst.png')
```

```python
from numpy import sin, cos, pi, sqrt

# OLYMPIC LOGO

N = 500
x = np.linspace(-10,10,N)
y = x
X,Y = np.meshgrid(x,y)
Rd = np.ones((N,N))
Gn = np.ones((N,N))
Bl = np.ones((N,N))

# draw colored circles
Rt,Rc,Ri,deg = 2,3,2.5,30

xt,yt = Rt*cos(deg*pi/180), Rt*sin(deg*pi/180)

# TOP MIDDLE (BLACK)
R = sqrt((X)**2 + (Y+Rt)**2)

Rd[np.where(R<Rc)] = 0.0    # RGB of BLACK ring
Gn[np.where(R<Rc)] = 0.0
Bl[np.where(R<Rc)] = 0.0

Rd[np.where(R<Ri)] = 1.0    # white inside
Gn[np.where(R<Ri)] = 1.0
Bl[np.where(R<Ri)] = 1.0

# TOP RIGHT (RED)
R = sqrt((X-4*xt)**2 + (Y+Rt)**2)

Rd[np.where(R<Rc)] = 1.0    # RGB of RED ring
Gn[np.where(R<Rc)] = 0.0
Bl[np.where(R<Rc)] = 0.0

Rd[np.where(R<Ri)] = 1.0    # white inside
Gn[np.where(R<Ri)] = 1.0
Bl[np.where(R<Ri)] = 1.0

# TOP LEFT (BLUE)
R = sqrt((X+4*xt)**2 + (Y+Rt)**2)

Rd[np.where(R<Rc)] = 0.0    # RGB of BLUE ring
Gn[np.where(R<Rc)] = 0.0
Bl[np.where(R<Rc)] = 1.0

Rd[np.where(R<Ri)] = 1.0    # white inside
Gn[np.where(R<Ri)] = 1.0
Bl[np.where(R<Ri)] = 1.0

# BOTTOM RIGHT (GREEN)
R = sqrt((X-2*xt)**2 + (Y-yt)**2)

Rd[np.where(R<Rc)] = 0.0    # RGB of GREEN ring
Gn[np.where(R<Rc)] = 1.0
Bl[np.where(R<Rc)] = 0.0

Rd[np.where(R<Ri)] = 1.0    # white inside
Gn[np.where(R<Ri)] = 1.0
Bl[np.where(R<Ri)] = 1.0

# BOTTOM LEFT (YELLOW)
R = sqrt((X+2*xt)**2 + (Y-yt)**2)

Rd[np.where(R<Rc)] = 1.0    # RGB of YELLOW ring
Gn[np.where(R<Rc)] = 1.0
Bl[np.where(R<Rc)] = 0.0

Rd[np.where(R<Ri)] = 1.0    # white inside
Gn[np.where(R<Ri)] = 1.0
Bl[np.where(R<Ri)] = 1.0

# OVERLAP TOP LEFT - BOTTOM LEFT (BLUE-YELLOW)
R1 = sqrt((X+4*xt)**2 + (Y+Rt)**2)
R2 = sqrt((X+2*xt)**2 + (Y-yt)**2)

for i in range(N):              # check if every pixel is between the gap of BLUE and YELLOW ring
    for j in range(N):
        if R1[i,j] < Rc and R2[i,j] < Ri:
            Rd[i,j] = 0.0
            Gn[i,j] = 0.0
            Bl[i,j] = 1.0

for i in range(N):              # set the actual gap between BLUE and YELLOW to white background
    for j in range(N):
        if R1[i,j] < Ri and R2[i,j] < Ri:
            Rd[i,j] = 1.0
            Gn[i,j] = 1.0
            Bl[i,j] = 1.0
            
# OVERLAP TOP MIDDLE - BOTTOM LEFT (BLACK-YELLOW)
R1 = sqrt((X)**2 + (Y+Rt)**2)
R2 = sqrt((X+2*xt)**2 + (Y-yt)**2)

for i in range(N):              # check if every pixel is between the gap of BLACK and YELLOW ring
    for j in range(N):
        if R1[i,j] < Rc and R2[i,j] < Ri:
            Rd[i,j] = 0.0
            Gn[i,j] = 0.0
            Bl[i,j] = 0.0

for i in range(N):              # set the actual gap between BLACK and YELLOW to white background
    for j in range(N):
        if R1[i,j] < Ri and R2[i,j] < Ri:
            Rd[i,j] = 1.0
            Gn[i,j] = 1.0
            Bl[i,j] = 1.0
            
# OVERLAP TOP MIDDLE - BOTTOM RIGHT (BLACK-GREEN)
R1 = sqrt((X)**2 + (Y+Rt)**2)
R2 = sqrt((X-2*xt)**2 + (Y-yt)**2)

for i in range(N):              # check if every pixel is between the gap of BLACK and GREEN ring
    for j in range(N):
        if R1[i,j] < Rc and R2[i,j] < Ri:
            Rd[i,j] = 0.0
            Gn[i,j] = 0.0
            Bl[i,j] = 0.0

for i in range(N):              # set the actual gap between BLACK and GREEN to white background
    for j in range(N):
        if R1[i,j] < Ri and R2[i,j] < Ri:
            Rd[i,j] = 1.0
            Gn[i,j] = 1.0
            Bl[i,j] = 1.0
            
# OVERLAP TOP RIGHT - BOTTOM RIGHT (RED-GREEN)
R1 = sqrt((X-4*xt)**2 + (Y+Rt)**2)
R2 = sqrt((X-2*xt)**2 + (Y-yt)**2)

for i in range(N):
    for j in range(N):
        if R1[i,j] < Rc and R2[i,j] < Ri:    # check if every pixel is between the gap of RED and GREEN ring
            Rd[i,j] = 1.0
            Gn[i,j] = 0.0
            Bl[i,j] = 0.0

for i in range(N):              # set the actual gap between RED and GREEN to white background
    for j in range(N):
        if R1[i,j] < Ri and R2[i,j] < Ri:
            Rd[i,j] = 1.0
            Gn[i,j] = 1.0
            Bl[i,j] = 1.0

I = np.zeros((N,N,3))

I[...,0] = Rd
I[...,1] = Gn
I[...,2] = Bl

fig = plt.figure()
plt.imshow(I)
# plt.imsave('olympics.jpg',I)
# plt.imsave('olympics.png',I)
# plt.imsave('olympics.bmp',I)
# plt.imsave('olympics.tiff',I)
```

```python
# HISTOGRAM BACKPROJECTION

import cv2
dark_img = cv2.imread('dark_image.jpg')

dark_img_gray = cv2.cvtColor(dark_img, cv2.COLOR_BGR2GRAY)

# cv2.imshow('sdfasf',dark_img_gray)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
x = np.arange(0,256)

dark_img_gray_arr = np.reshape(dark_img_gray, (-1,1))

dark_img_gray_PDF = plt.hist(dark_img_gray_arr, bins=x, density=True)    # normalized histogram
plt.show()

cumsum = np.cumsum(dark_img_gray_PDF[0])
cumsum_arr = np.reshape(cumsum, (-1,1))
dark_img_gray_CDF = plt.plot(cumsum_arr)
plt.show()
```
