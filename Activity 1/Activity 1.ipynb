```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin, cos, pi, sqrt

# SINUSOID

N = 400

x = np.linspace(-1,1,N)
y = sin(8*pi*x)    # sin(2k*pi x) k determines frequency per unit

X,Y = np.meshgrid(x,y)
sinusoid = np.transpose(Y)

plt.imshow(sinusoid, cmap='gray')
# plt.imsave('sinusoid.jpg',sinusoid,cmap='gray')
# plt.imsave('sinusoid.png',sinusoid,cmap='gray')
# plt.imsave('sinusoid.bmp',sinusoid,cmap='gray')
# plt.imsave('sinusoid.tiff',sinusoid,cmap='gray')
```

```python
# GRATING

x = np.ones(N)
x *= -1

ctr = 0
div = 10
    
for j in range(int(N/div)):
    for i in range(ctr*div,N):
        x[i] *= -1
    ctr += 1

y = x
        
X,Y = np.meshgrid(x,y)

plt.imshow(X, cmap='gray')
# plt.imsave('gratings.jpg',X,cmap='gray')
# plt.imsave('gratings.png',X,cmap='gray')
# plt.imsave('gratings.bmp',X,cmap='gray')
# plt.imsave('gratings.tiff',X,cmap='gray')
```

```python
# ANNULUS

N = 2000 # the higher num is the finer
x = np.linspace(-1,1,num = N)
y = x
X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2)
A = np.zeros(np.shape(R))
A[np.where(R<0.9)] = 1.0       # set radius of 0.9 to 1 to create a disk
A[np.where(R<0.2)] = 0.0       # set radius of 0.2 back to 0 to punch a hole in the middle

plt.imshow(A, cmap='gray_r')
# plt.imsave('annulus.jpg',A,cmap='gray')
# plt.imsave('annulus.png',A,cmap='gray')
# plt.imsave('annulus.bmp',A,cmap='gray')
# plt.imsave('annulus.tiff',A,cmap='gray')
```

```python
# JWST

def space(plots, occupied):
    coord = plots        
    colors = ["white" if k == 0 else "yellow" for k in occupied.values()]
    
    # Horizontal coordinates array
    hcoord = [c[0] for c in coord]                              

    # Vertical coordinates array
    vcoord = [2.*sin(radians(60))*(c[1]-c[2])/3. for c in coord]
                                            

    fig, ax = plt.subplots(1, figsize=(8,8))
    ax.set_aspect('equal')                  

    
    for i, j, k in zip(hcoord, vcoord, colors):
        color = k[0]                                                       
        hex = RegularPolygon((i, j), numVertices=6, radius=2./3,           
                             orientation=radians(30), facecolor = color,   
                             alpha=0.3, edgecolor='k')
        ax.add_patch(hex)    


    ax.scatter(hcoord, vcoord, alpha=0)
    
    plt.axis('off')
    plt.show()
#     plt.savefig('jwst.png')
```

```python
from numpy import sin, cos, pi, sqrt

# OLYMPIC LOGO

N = 500
x = np.linspace(-10,10,N)
y = x
X,Y = np.meshgrid(x,y)
Rd = np.ones((N,N))
Gn = np.ones((N,N))
Bl = np.ones((N,N))

# draw colored circles
Rt,Rc,Ri,deg = 2,3,2.5,30

xt,yt = Rt*cos(deg*pi/180), Rt*sin(deg*pi/180)

# TOP MIDDLE (BLACK)
R = sqrt((X)**2 + (Y+Rt)**2)

Rd[np.where(R<Rc)] = 0.0    # RGB of BLACK ring
Gn[np.where(R<Rc)] = 0.0
Bl[np.where(R<Rc)] = 0.0

Rd[np.where(R<Ri)] = 1.0    # white inside
Gn[np.where(R<Ri)] = 1.0
Bl[np.where(R<Ri)] = 1.0

# TOP RIGHT (RED)
R = sqrt((X-4*xt)**2 + (Y+Rt)**2)

Rd[np.where(R<Rc)] = 1.0    # RGB of RED ring
Gn[np.where(R<Rc)] = 0.0
Bl[np.where(R<Rc)] = 0.0

Rd[np.where(R<Ri)] = 1.0    # white inside
Gn[np.where(R<Ri)] = 1.0
Bl[np.where(R<Ri)] = 1.0

# TOP LEFT (BLUE)
R = sqrt((X+4*xt)**2 + (Y+Rt)**2)

Rd[np.where(R<Rc)] = 0.0    # RGB of BLUE ring
Gn[np.where(R<Rc)] = 0.0
Bl[np.where(R<Rc)] = 1.0

Rd[np.where(R<Ri)] = 1.0    # white inside
Gn[np.where(R<Ri)] = 1.0
Bl[np.where(R<Ri)] = 1.0

# BOTTOM RIGHT (GREEN)
R = sqrt((X-2*xt)**2 + (Y-yt)**2)

Rd[np.where(R<Rc)] = 0.0    # RGB of GREEN ring
Gn[np.where(R<Rc)] = 1.0
Bl[np.where(R<Rc)] = 0.0

Rd[np.where(R<Ri)] = 1.0    # white inside
Gn[np.where(R<Ri)] = 1.0
Bl[np.where(R<Ri)] = 1.0

# BOTTOM LEFT (YELLOW)
R = sqrt((X+2*xt)**2 + (Y-yt)**2)

Rd[np.where(R<Rc)] = 1.0    # RGB of YELLOW ring
Gn[np.where(R<Rc)] = 1.0
Bl[np.where(R<Rc)] = 0.0

Rd[np.where(R<Ri)] = 1.0    # white inside
Gn[np.where(R<Ri)] = 1.0
Bl[np.where(R<Ri)] = 1.0

# OVERLAP TOP LEFT - BOTTOM LEFT (BLUE-YELLOW)
R1 = sqrt((X+4*xt)**2 + (Y+Rt)**2)
R2 = sqrt((X+2*xt)**2 + (Y-yt)**2)

for i in range(N):              # check if every pixel is between the gap of BLUE and YELLOW ring
    for j in range(N):
        if R1[i,j] < Rc and R2[i,j] < Ri:
            Rd[i,j] = 0.0
            Gn[i,j] = 0.0
            Bl[i,j] = 1.0

for i in range(N):              # set the actual gap between BLUE and YELLOW to white background
    for j in range(N):
        if R1[i,j] < Ri and R2[i,j] < Ri:
            Rd[i,j] = 1.0
            Gn[i,j] = 1.0
            Bl[i,j] = 1.0
            
# OVERLAP TOP MIDDLE - BOTTOM LEFT (BLACK-YELLOW)
R1 = sqrt((X)**2 + (Y+Rt)**2)
R2 = sqrt((X+2*xt)**2 + (Y-yt)**2)

for i in range(N):              # check if every pixel is between the gap of BLACK and YELLOW ring
    for j in range(N):
        if R1[i,j] < Rc and R2[i,j] < Ri:
            Rd[i,j] = 0.0
            Gn[i,j] = 0.0
            Bl[i,j] = 0.0

for i in range(N):              # set the actual gap between BLACK and YELLOW to white background
    for j in range(N):
        if R1[i,j] < Ri and R2[i,j] < Ri:
            Rd[i,j] = 1.0
            Gn[i,j] = 1.0
            Bl[i,j] = 1.0
            
# OVERLAP TOP MIDDLE - BOTTOM RIGHT (BLACK-GREEN)
R1 = sqrt((X)**2 + (Y+Rt)**2)
R2 = sqrt((X-2*xt)**2 + (Y-yt)**2)

for i in range(N):              # check if every pixel is between the gap of BLACK and GREEN ring
    for j in range(N):
        if R1[i,j] < Rc and R2[i,j] < Ri:
            Rd[i,j] = 0.0
            Gn[i,j] = 0.0
            Bl[i,j] = 0.0

for i in range(N):              # set the actual gap between BLACK and GREEN to white background
    for j in range(N):
        if R1[i,j] < Ri and R2[i,j] < Ri:
            Rd[i,j] = 1.0
            Gn[i,j] = 1.0
            Bl[i,j] = 1.0
            
# OVERLAP TOP RIGHT - BOTTOM RIGHT (RED-GREEN)
R1 = sqrt((X-4*xt)**2 + (Y+Rt)**2)
R2 = sqrt((X-2*xt)**2 + (Y-yt)**2)

for i in range(N):
    for j in range(N):
        if R1[i,j] < Rc and R2[i,j] < Ri:    # check if every pixel is between the gap of RED and GREEN ring
            Rd[i,j] = 1.0
            Gn[i,j] = 0.0
            Bl[i,j] = 0.0

for i in range(N):              # set the actual gap between RED and GREEN to white background
    for j in range(N):
        if R1[i,j] < Ri and R2[i,j] < Ri:
            Rd[i,j] = 1.0
            Gn[i,j] = 1.0
            Bl[i,j] = 1.0

I = np.zeros((N,N,3))

I[...,0] = Rd
I[...,1] = Gn
I[...,2] = Bl

fig = plt.figure()
plt.imshow(I)
# plt.imsave('olympics.jpg',I)
# plt.imsave('olympics.png',I)
# plt.imsave('olympics.bmp',I)
# plt.imsave('olympics.tiff',I)
```

```python
# HISTOGRAM BACKPROJECTION
def mean_gray(img):
    grayscale = np.dot(img[...,:3],[0.2989, 0.5870, 0.1140])
    return grayscale

dark_img = plt.imread('dark_image.jpg')
dark_img_gray = mean_gray(dark_img)

plt.imshow(dark_img_gray,cmap='gray')
```

```python
dark_img_gray_norm = 255*dark_img_gray/dark_img_gray.max()

plt.figure(figsize=(15,5))

plt.subplot(121)
dark_img_gray_PDF, bins = np.histogram(dark_img_gray_norm, bins=256, density=True)    # normalized histogram
plt.xlim(0,255)
plt.title('PDF of original image')
plt.plot(bins[:-1], dark_img_gray_PDF)

cumsum = np.cumsum(dark_img_gray_PDF)
plt.subplot(122)
plt.xlim(0,255)
plt.title('CDF of riginal image')
dark_img_gray_CDF = plt.plot(bins[:-1],cumsum)

plt.show()
```

```python
# desired CDF (Linear)

desired_x = np.linspace(0,256,256)
desired_y = np.linspace(0,1,256)

plt.title("Desired CDF (Linear)")
plt.xlim(0,255)
plt.ylim(0,1)
plt.plot(desired_x, desired_y)
plt.show()
```

```python
dark_img_gray_int = dark_img_gray.astype(int)

old_x = desired_x
new_x = cumsum*255

dark_img_gray_flat = dark_img_gray_int.ravel()

index = np.searchsorted(old_x, dark_img_gray_flat)

dark_img_gray_flat[:] = new_x[index]
```

```python
plt.figure(figsize=(15,5))

plt.subplot(121)
plt.title('Original image')
plt.imshow(dark_img_gray, cmap='gray')

plt.subplot(122)
plt.title('Histogram equalized image (Linear)')
plt.imshow(dark_img_gray_int, cmap = 'gray')

plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(121)
dark_img_eqlzd_PDF, bins = np.histogram(dark_img_gray_int, bins=256, density=True)    # normalized histogram
plt.xlim(0,255)
plt.title('PDF of histogram equalized image')
plt.plot(bins[:-1], dark_img_eqlzd_PDF)

cumsum = np.cumsum(dark_img_eqlzd_PDF)
plt.subplot(122)
plt.xlim(0,255)
plt.title('CDF of histogram equalized image')
dark_img_gray_CDF = plt.plot(bins[:-1],cumsum)

plt.show()
```

```python
# desired CDF (Sigmoid)

def sigmoid(x):
    return 1 / (1 + np.exp((med-x)/15))

def inverse_sigmoid(y):
    return med - 15*np.log(abs((1/y)-1))

sigmoid_x = np.arange(0,256)
med = np.median(sigmoid_x)
sigmoid_y = sigmoid(sigmoid_x)

plt.title("Desired CDF (Sigmoid)")
plt.xlim(0,255)
plt.ylim(0,1)
plt.plot(sigmoid_x, sigmoid_y)
plt.show()
```

```python
dark_img_gray_norm = 255*dark_img_gray/dark_img_gray.max()

dark_img_gray_PDF, bins = np.histogram(dark_img_gray_norm, bins=256, density=True)    # normalized histogram

cumsum = np.cumsum(dark_img_gray_PDF)
```

```python
dark_img_gray_int = dark_img_gray.astype(int)

old_x = desired_x
new_x = inverse_sigmoid(cumsum)

mask = np.in1d(dark_img_gray_int,old_x)

index = np.searchsorted(old_x,dark_img_gray_int.ravel()[mask])

dark_img_gray_int.ravel()[mask] = new_x[index]
```

```python
# Same plotting parameters
plt.figure(figsize = (15,5))


plt.subplot(121)
plt.title("Original image")
plt.imshow(dark_img_gray, cmap = 'gray')

plt.subplot(122)
plt.title('Histogram equalized image (Sigmoid)')
plt.imshow(dark_img_gray_int, cmap = 'gray')

plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(121)
dark_img_eqlzd_PDF, bins = np.histogram(dark_img_gray_int, bins=256, density=True)    # normalized histogram
plt.title('PDF of histogram equalized image')
plt.plot(bins[:-1], dark_img_eqlzd_PDF)

cumsum = np.cumsum(dark_img_eqlzd_PDF)
plt.subplot(122)
plt.title('CDF of histogram equalized image')
dark_img_eqlzd_CDF = plt.plot(desired_x,cumsum)

plt.show()
```

```python
# CONTRAST STRECTCHING
def mean_gray(img):
    grayscale = np.dot(img[...,:3],[0.2989, 0.5870, 0.1140])
    return grayscale

dark_img = plt.imread('dark_image.jpg')
dark_img_gray = mean_gray(dark_img)

plt.imshow(dark_img_gray,cmap='gray')
```

```python
dark_img_gray_norm = 255*dark_img_gray/dark_img_gray.max()

plt.figure(figsize=(15,5))

plt.subplot(121)
dark_img_gray_PDF, bins = np.histogram(dark_img_gray_norm, bins=256, density=True)    # normalized histogram
plt.xlim(0,255)
plt.title('PDF of original image')
plt.plot(bins[:-1], dark_img_gray_PDF)

dark_img_gray_CDF = np.cumsum(dark_img_gray_PDF)
plt.subplot(122)
plt.xlim(0,255)
plt.title('CDF of riginal image')
plt.plot(bins[:-1],dark_img_gray_CDF)

plt.show()
```

```python
def contrast_stretch(img,min_percentile,max_percentile):
    img_gray = mean_gray(img)
    img_gray_norm = 255*img_gray/img_gray.max()
    img_gray_PDF, bins = np.histogram(img_gray, bins=256, density=True)
    img_gray_CDF = np.cumsum(img_gray_PDF)
    
    mins = np.where(img_gray_CDF <= min_percentile)
    I_min = mins[0].max()
    
    maxs = np.where(img_gray_CDF >= max_percentile)
    I_max = maxs[0].min()
    
    
    I = (img_gray - I_min)/(I_max - I_min)
    return I*255
    
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.imshow(dark_img_gray,cmap='gray')

dark_img_gray_stretched = contrast_stretch(dark_img,0.2,0.8)
plt.subplot(122)
plt.imshow(dark_img_gray_stretched,cmap='gray')
```
