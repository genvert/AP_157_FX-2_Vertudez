```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin, cos, pi, sqrt

N = 400

x = np.linspace(-1,1,N)
y = sin(8*pi*x)

X,Y = np.meshgrid(x,y)

plt.imshow(np.transpose(Y), cmap='gray')
```

```python
x = np.ones(N)
x *= -1

ctr = 0
div = 10
    
for j in range(int(N/div)):
    for i in range(ctr*div,N):
        x[i] *= -1
    ctr += 1

y = x
        
X,Y = np.meshgrid(x,y)

plt.imshow(X, cmap='gray')
```

```python
N = 2000 # the higher num is the finer
x = np.linspace(-1,1,num = N)
y = x
X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2)
A = np.zeros(np.shape(R))
A[np.where(R<0.9)] = 1.0       # set radius of 0.9 to 1 to create a disk
A[np.where(R<0.2)] = 0.0       # set radius of 0.2 back to 0 to punch a hole in the middle

plt.imshow(A, cmap='gray_r')
```

```python
N = 1500 #the higher num is the finer
x = np.linspace(-1,1,num = N)
y = x

X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2)
A = np.zeros(np.shape(R))

corners = (2/1500)*sqrt((125**2)+( (125*(sqrt(3) - 1))/(sqrt(3)) )**2)

A[np.where(R == corners)] = 1.0

plt.imshow(A, cmap='YlOrBr')
```

```python
from numpy import sin, cos, pi, sqrt

N = 500
x = np.linspace(-10,10,N)
y = x
X,Y = np.meshgrid(x,y)
Rd = np.zeros((N,N))
Gn = np.zeros((N,N))
Bl = np.zeros((N,N))
```

```python
# draw colored circles
Rt,Rc,deg = 3,4,30

xt,yt = Rt*cos(deg*pi/180), Rt*sin(deg*pi/180)

R = sqrt((X)**2 + (Y+Rt)**2)
Rd[np.where(R<Rc)] = 1.0
Rd[np.where(R<3)] = 0.0

R = sqrt((X-2*xt)**2 + (Y+Rt)**2)
Rd[np.where(R<Rc)] = 1.0
Rd[np.where(R<3)] = 0.0

R = sqrt((X+2*xt)**2 + (Y+Rt)**2)
Rd[np.where(R<Rc)] = 1.0
Rd[np.where(R<3)] = 0.0

R = sqrt((X-xt)**2 + (Y-yt)**2)
Gn[np.where(R<Rc)] = 1.0
Rd[np.where(R<3)] = 0.0

R = sqrt((X+xt)**2 + (Y-yt)**2)
Gn[np.where(R<Rc)] = 1.0
Rd[np.where(R<Rc)] = 1.0
Rd[np.where(R<3)] = 0.0

I = np.zeros((N,N,3))

I[...,0] = Rd
I[...,1] = Gn
I[...,2] = Bl

fig = plt.figure()
plt.imshow(I)
```
