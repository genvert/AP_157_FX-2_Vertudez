```python
import numpy as np
import matplotlib.pyplot as plt

N = 200 #the higher num is the finer
x = np.linspace(-1,1,num = N)
y = x
X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2)
A = np.zeros(np.shape(R))
A[np.where(R<0.5)] = 1.0

plt.imshow(A, cmap='gray')

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X,Y,A)
```

```python
from numpy import sin, cos, pi, sqrt

N = 500
x = np.linspace(-10,10,N)
y = x
X,Y = np.meshgrid(x,y)
Rd, Gn, Bl = np.zeros((N,N)),np.zeros((N,N)),np.zeros((N,N))
```

```python
# draw colored circles
Rt,Rc,deg = 3,4,30

xt,yt = Rt*cos(deg*pi/180), Rt*sin(deg*pi/180)

R = sqrt((X)**2 + (Y+Rt)**2)
Rd[np.where(R<Rc)] = 1.0

R = sqrt((X-xt)**2 + (Y-yt)**2)
Gn[np.where(R<Rc)] = 1.0

R = sqrt((X+xt)**2 + (Y-yt)**2)
Bl[np.where(R<Rc)] = 1.0

I = np.zeros((N,N,3))

I[...,0] = Rd
I[...,1] = Gn
I[...,2] = Bl

fig = plt.figure()
plt.imshow(I)
```
