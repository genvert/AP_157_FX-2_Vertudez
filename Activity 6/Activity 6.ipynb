```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def subplots(img1,img2,img3,title1='Original image',title2='Structuring element',title3=''):
    plt.figure(figsize=(15,5))
    
    plt.subplot(131)
    plt.imshow(img1,cmap='viridis')
    plt.title(title1)
    
    plt.subplot(132)
    plt.imshow(img2,cmap='viridis_r')
    plt.title(title2)
    
    plt.subplot(133)
    plt.imshow(img3, cmap='viridis')
    plt.title(title3)
    
    plt.show()
```

# Morphological operations

```python
SE_a = np.ones([2,2])
SE_b = np.ones([1,2])
SE_c = np.ones([5,5])
```

```python
square = np.zeros([11,11])
square[3:8,3:8] = 1

square_dilated_a = cv2.dilate(square,SE_a,anchor=(0,0))
square_eroded_a = cv2.erode(square,SE_a,anchor=(0,0))

square_dilated_b = cv2.dilate(square,SE_b,anchor=(0,0))
square_eroded_b = cv2.erode(square,SE_b,anchor=(0,0))

square_dilated_c = cv2.dilate(square,SE_c,anchor=(-1,-1))
square_eroded_c = cv2.erode(square,SE_c,anchor=(-1,-1))

subplots(square,SE_a,square_dilated_a,title3='Dilated image')
subplots(square,SE_a,square_eroded_a,title3='Eroded image')
subplots(square,SE_b,square_dilated_b,title3='Dilated image')
subplots(square,SE_b,square_eroded_b,title3='Eroded image')
subplots(square,SE_c,square_dilated_c,title3='Dilated image')
subplots(square,SE_c,square_eroded_c,title3='Eroded image')
```

```python
hollow = np.zeros([20,20])
hollow[5:15,5:15] = 1
hollow[7:13,7:13] = 0

hollow_dilated_a = cv2.dilate(hollow,SE_a,anchor=(0,0))
hollow_eroded_a = cv2.erode(hollow,SE_a,anchor=(0,0))

hollow_dilated_b = cv2.dilate(hollow,SE_b,anchor=(0,0))
hollow_eroded_b = cv2.erode(hollow,SE_b,anchor=(0,0))

hollow_dilated_c = cv2.dilate(hollow,SE_c,anchor=(-1,-1))
hollow_eroded_c = cv2.erode(hollow,SE_c,anchor=(-1,-1))

subplots(hollow,SE_a,hollow_dilated_a,title3='Dilated image')
subplots(hollow,SE_a,hollow_eroded_a,title3='Eroded image')
subplots(hollow,SE_b,hollow_dilated_b,title3='Dilated image')
subplots(hollow,SE_b,hollow_eroded_b,title3='Eroded image')
subplots(hollow,SE_c,hollow_dilated_c,title3='Dilated image')
subplots(hollow,SE_c,hollow_eroded_c,title3='Eroded image')
```
